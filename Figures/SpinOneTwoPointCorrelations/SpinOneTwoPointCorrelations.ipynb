# Spin-one two-point correlations

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import pickle

#%matplotlib inline
```

Load data

```python
with open('../../Data/CorrSpin1L1001N1000tbeta1h0.pickle', 'rb') as pfile:
    structure = pickle.load(pfile)

corr_ninj1=structure['corr_ninj1F']
N=structure['NF']
L=structure['LF']
h=structure['hF']
Nl=int((L)/2)
```

```python
with open('../../Data/CorrSpin1L1001N1000tbeta1h5s1s1.pickle', 'rb') as pfile:
    structure = pickle.load(pfile)

corr_ninj1h11=structure['corr_ninj1h11F']
N=structure['NF']
L=structure['LF']
h=structure['hF']
Nl=int((L)/2)
```

```python
with open('../../Data/CorrSpin1L1001N1000tbeta1h5s1s0.pickle', 'rb') as pfile:
    structure = pickle.load(pfile)

corr_ninj1h10=structure['corr_ninj1h10F']
N=structure['NF']
L=structure['LF']
h=structure['hF']
Nl=int((L)/2)
```

```python
with open('../../Data/CorrSpin1L1001N1000tbeta1h5s1s-1.pickle', 'rb') as pfile:
    structure = pickle.load(pfile)

corr_ninj1h1neg1=structure['corr_ninj1h1neg1F']
N=structure['NF']
L=structure['LF']
h=structure['hF']
Nl=int((L)/2)
```

Generate the figure

```python
Index=np.arange(-Nl,Nl+1)
with plt.style.context('../aps2.mplstyle'):
    fig1, ax1 = plt.subplots(2,2,figsize=[246.0*1.0/72.27 /.8,246.0*1.0/72.27 ])
    Tcorr_ninj1=np.transpose(corr_ninj1) 
    im1 = ax1[0,0].pcolormesh(Index,Index,-Tcorr_ninj1,rasterized=True,norm=colors.LogNorm())
    cb00=fig1.colorbar(im1,ax=ax1[0,0],ticks=[1.0e-3, 1.0e-5, 1.0e-7],pad=0.03)
    cb00.ax.set_yticklabels([r'$-10^{-3}$',r'$-10^{-5}$', r'$-10^{-7}$']) 
    ax1[0,0].yaxis.set_label_coords(-0.20, 0.5)
    ax1[0,0].set_ylabel(r'$j$')
    #ax1[0,0].set_xticklabels([])
    ax1[0,0].set_title(r'$h=0.0$')
    ax1[0,0].tick_params(axis='x', which='bottom', length=0)
    ax1[0,0].tick_params(axis='x', labelbottom="off", labeltop="off")


    Tcorr_ninj1h11=np.transpose(corr_ninj1h11) 
    im1 = ax1[1,0].pcolormesh(Index,Index,-Tcorr_ninj1h11,rasterized=True,norm=colors.LogNorm())
    cb10=fig1.colorbar(im1,ax=ax1[1,0],ticks=[1.0e-1,1.0e-3, 1.0e-5, 1.0e-7],pad=0.03)
    cb10.ax.set_yticklabels([r'$-10^{-1}$',r'$-10^{-3}$',r'$-10^{-5}$', r'$-10^{-7}$'])
    #cb10.ax.yaxis.set_label_coords(3.0, .70)    
    ax1[1,0].yaxis.set_label_coords(-0.20, 0.5)
    ax1[1,0].xaxis.set_label_coords( 0.5,-0.20)
    ax1[1,0].set_xlabel(r'$i$')
    ax1[1,0].set_ylabel(r'$j$')
    ax1[1,0].set_title(r'$h=5t,\sigma=1,\sigma^\prime=1$')

    Tcorr_ninj1h10=np.transpose(corr_ninj1h10) 
    im1 = ax1[0,1].pcolormesh(Index,Index,-Tcorr_ninj1h10,rasterized=True,norm=colors.LogNorm())
    cb01=fig1.colorbar(im1,ax=ax1[0,1],ticks=[1.0e-5, 1.0e-7, 1.0e-9],pad=0.03,label=r'$\mathcal{C}(n_{i,\sigma},n_{j,\sigma^\prime})$')
    cb01.ax.set_yticklabels([r'$-10^{-5}$',r'$-10^{-7}$',r'$-10^{-9}$'])  
    ax1[0,1].set_xticklabels([])
    ax1[0,1].set_yticklabels([])
    ax1[0,1].set_title(r'$h=5t,\sigma=1,\sigma^\prime=0$')
    ax1[0,1].tick_params(axis='y', which='both', length=0)

    Tcorr_ninj1h1neg1=np.transpose(corr_ninj1h1neg1) 
    im1 = ax1[1,1].pcolormesh(Index,Index,-Tcorr_ninj1h1neg1,rasterized=True,norm=colors.LogNorm())
    cb11=fig1.colorbar(im1,ax=ax1[1,1],ticks=[1.0e-11, 1.0e-12, 1.0e-13],pad=0.03,label=r'$\mathcal{C}(n_{i,\sigma},n_{j,\sigma^\prime})$')
    cb11.ax.set_yticklabels([r'$-10^{-11}$',r'$-10^{-12}$',r'$-10^{-13}$'])
    ax1[1,1].set_xlabel(r'$i$')
    ax1[1,1].set_yticklabels([])
    ax1[1,1].set_title(r'$h=5t,\sigma=0,\sigma^\prime=-1$')
    ax1[1,1].tick_params(axis='y', which='both', length=0)
    ax1[1,1].xaxis.set_label_coords( 0.5,-0.20)
    ax1[0,0].set_aspect(1)
    ax1[1,0].set_aspect(1)
    ax1[0,1].set_aspect(1)
    ax1[1,1].set_aspect(1)

    plt.tight_layout(h_pad=1)
    # Adjust space between subplots
    plt.subplots_adjust(hspace=0.30)

    #Adjust space between subplots
    plt.subplots_adjust(wspace = 0.30)
    
    #plt.savefig('CorrSpin1L1001N1000tbeta1.pdf', transparent=False)  
    #plt.savefig('CorrSpin1L1001N1000tbeta1.svg', transparent=False)  
    plt.show()
```
